#Predominant age group among employees

```python
age_group= df.Age.value_counts()
age_group
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x= 'Age', data= df)
plt.title("")
```

```python
df['Age Group'] = df['Age'].apply(lambda Age:
                               '20-25' if (20<=Age<=25) else
                                    ('26-30' if (26<=Age<=30) else
                                        ('31-35' if (31<=Age<=35) else
                                             ('36-40' if (36<=Age<=40) else
                                                 ('40 above' if(41>=Age) else 'other')))))
df
```

```python
age_group= df['Age Group'].value_counts()
age_group
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Age Group', data= df, hue='Age Group', palette='dark')
```

Data Story: The countplot analysis of employee age distribution reveals that the majority of employees, approximately 198 individuals, fall within the 20-25 age range. In contrast, there are fewer employees, with only 25 individuals, in the 36-40 age range.

#Highest Salary Expenditure

```python
df.Salary.value_counts()
```

```python
total_salary= df.groupby(['Team', 'Position']).Salary.sum()
total_salary
```

```python
max_spending_team= total_salary.idxmax()
max_spending_team
```

```python
max_spending_salary= total_salary.max()
max_spending_salary
```

```python
print(f"Highest spending team and position:\n Team: {max_spending_team[0]} \n Position: {max_spending_team[1]}")
print(f"And the Highest salary expenditure is {max_spending_salary}")
```

#scatterplot:

```python
plt.figure(figsize=(25,6))
sns.scatterplot(x= 'Team', y= 'Salary', data= df, hue= 'Position')
plt.title("Highest Salary Expenditure")
plt.xticks(rotation=45)
plt.show()
```

The scatter plot illustrates the salary distribution across various positions within each team,
revealing that the Los Angeles Lakers stand out as the highest-spending team.

#Correlation Between Age and Salary

```python
crr_df= df[['Age', 'Salary']]
crr_df
```

```python
corr= crr_df.corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='viridis')
plt.title("Correlation matrix between age and salary of employees")
plt.show()
```

Age vs Age(1): This indicates a perfect positive correlation.
Age vs. Salary (0.21117): The correlation coefficient of 0.21117 indicates a positive, yet relatively weak, relationship between age and salary. This suggests that as age increases, there is a slight tendency for salaries to rise, although the correlation is not very strong.
The correlation matrix reveals a positive, though not particularly strong relationship between age and salary in the employee dataset.
As employee's ages increase, there is a slight tendency for their salaries to rise accordingly.
